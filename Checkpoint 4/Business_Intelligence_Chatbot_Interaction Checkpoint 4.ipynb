```python
pd.DataFrame(df.isna().sum() / len(df) * 100, columns=['Null Ratio in %'])
```

1. **user_feedback_rating**

Kolom ini bertipe numerik (skala 1-5). Karena distribusinya simetris (Mean $\approx$ Median), pengisian dilakukan menggunakan nilai Median. Hal ini dilakukan untuk menghindari bias yang mungkin muncul jika menggunakan Mean apabila terdapat nilai ekstrem yang tidak terlihat.

```python
df['user_feedback_rating'] = df['user_feedback_rating'].fillna(df['user_feedback_rating'].dropna().median())
```

2. **estimated_business_impact**

Kolom ini bertipe kategorikal (High, Medium, Low). Karena data bersifat label, maka pengisian dilakukan menggunakan Modus (kategori yang paling sering muncul). Ini memastikan integritas kategori tetap terjaga tanpa menambah label baru yang tidak terdefinisi.

```python
df['estimated_business_impact'] = df['estimated_business_impact'].fillna(df['estimated_business_impact'].dropna().mode()[0])
```

Dengan persentase 0%. Tingkat kepercayaan bot tersebar secara konsisten dalam rentang yang wajar, menandakan tidak ada jawaban yang memiliki skor kepercayaan yang terlalu rendah atau terlalu tinggi secara tidak wajar.

* user_feedback_rating

Dengan persentase 0%. Penilaian pengguna berada sepenuhnya dalam skala 1 hingga 5 tanpa adanya nilai di luar batas tersebut, sehingga data persepsi kepuasan pengguna dianggap stabil.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

# **Data Construction**

Pada tahap ini, saya melakukan konstruksi data untuk menciptakan variabel baru yang dapat memberikan insight lebih mendalam mengenai performa sistem. Saya membuat kolom response_category untuk mengklasifikasikan waktu respon chatbot ke dalam kategori yang lebih mudah dipahami oleh manajemen.

Alasan:
Data dalam milidetik (response_time_ms) bersifat terlalu teknis. Dengan mengategorikannya menjadi 'Fast', 'Average', dan 'Slow', kita dapat langsung memetakan distribusi efisiensi sistem dan melihat apakah ada korelasi antara respon yang lambat dengan penurunan kepuasan pengguna.

```python
def kategori_respon(waktu):
    if waktu <= 1000:
        return 'Fast'
    elif waktu <= 2000:
        return 'Average'
    else:
        return 'Slow'

df['response_category'] = df['response_time_ms'].apply(kategori_respon)
```
